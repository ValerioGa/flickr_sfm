```python
from pathlib import Path
import random
import os
import matplotlib.pyplot as plt
from PIL import Image
```

```python
def retrieval_pairs(file, im):

  pairs = []
  with open(file, "r") as f:
    for row in f:
      names = row.split()

      if names[0] == im:
        pair = names[1]
        pairs.append(pair)

  return pairs

#Path to the pairs file
pairs = '/outputs/torino/cosplace10_min0.0_skip0+superpoint_aachen+superglue/pairs.txt'

#Path to the train images folder
images = Path('/flickr_images/torino/train')
```

```python
#Name of the query image
query_im = '@0396328.82@4991660.49@32@T@045.07081@0007.68300@28566721038@@@@@@20180529@@.jpg'
```

```python
#Random query image if no specific one is needed
query_im = random.choice(os.listdir(images))
```

```python
pairs_retrieval = retrieval_pairs(pairs, query_im)

print("Searched image:")
image = Image.open(images/query_im)
plt.imshow(image)
plt.show()

print(f"Retrieved images: {len(pairs_retrieval)}")

for pair in pairs_retrieval:
  
  image = Image.open(images/pair)
  plt.imshow(image)
  plt.show()
```
